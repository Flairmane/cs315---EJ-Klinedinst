```python
import random
import math
```

```python
p = 23
q = 29

n = p*q
phi = (p-1)*(q-1)
```

```python
e = random.randrange(1, phi)
while(math.gcd(e, phi) != 1):
  e = random.randrange(1, phi)
print("The computed value of e is ", e)
```

```python
def modinverse(e, phi):
  e = e % phi
  for x in range(1, phi):
    if ((e * x) % phi == 1):
      return x
  return 1
```

```python
d = modinverse(e,phi)
print("d =", d)
print("The public key ğ‘ƒ = ( ğ‘’ , ğ‘› ) is ğ‘ƒ = (", e, ",", phi, ")")
print("The public key S = ( ğ‘’ , ğ‘› ) is S = (", d, ",", phi, ")")
```
